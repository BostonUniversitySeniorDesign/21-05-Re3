# make distribution plot of people that know R versus people that don't

```python
df2 = pd.read_csv('../data/matrix_of_ratings.csv')
df2 = df2.drop(df2.columns[[0]], axis=1)
df2.head()

ratings_with_exp = np.zeros((100,2))
ratings_no_exp = np.zeros((100,2))

for i in range(df2.shape[1] -1):
    for j in range(df2.shape[0] ):
        if(df["user" + str(i)][3] == 'Yes' and df2["user" + str(i)][j] > 0):
            ratings_with_exp[j,0] += df2["user" + str(i)][j]
            ratings_with_exp[j,1] += 1

        elif(df2["user" + str(i)][j] > 0):
            ratings_no_exp[j,0] += df2["user" + str(i)][j]
            ratings_no_exp[j,1] += 1

avg_ratings_with_exp = ratings_with_exp[:,0]/ratings_with_exp[:,1]

avg_ratings_no_exp = ratings_no_exp[:,0]/ratings_no_exp[:,1]

print(avg_ratings_with_exp)
print(avg_ratings_no_exp)

absolute_diff = 0
for i in range(100):
    absolute_diff += (abs(avg_ratings_with_exp[i] - avg_ratings_no_exp[i])/100)
    
print('''This is the absolute difference between the avg ratings of people
with R compare to people with no R experience: ''' + str(absolute_diff))




    
```
