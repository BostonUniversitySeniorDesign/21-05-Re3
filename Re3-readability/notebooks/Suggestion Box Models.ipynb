# Making simple regression models with features that correlate well

```python
list_of_strong_features = ['avg_line_len','avg_assignments','avg_parentheses','avg_periods']
```

```python
for i in list_of_strong_features:
    #initiating model
    linear_regression_model = LinearRegression()
    
    #getting the feature we are using it and shaping it so it can be passed to a model
    training_arr = np.reshape(features[i].values,(100,1))
    
    
    #fitting the model
    reg = linear_regression_model.fit(training_arr, ratings['snippet rating'].values) 

    #make prediction to evaluate R^2 and mse
    prediction = reg.predict(training_arr)
    
    res = stats.linregress(reg.predict(training_arr),ratings['snippet rating'].values)
    print(f"R-squared: {res.rvalue**2:.6f}")
    print("This is the mean squared error: "  + str(mean_squared_error(prediction,ratings['snippet rating'].values)))
    
    # save the model to disk
    filename = 'linear_model_' + i + '.sav'
    pickle.dump(reg, open(filename, 'wb'))
```
